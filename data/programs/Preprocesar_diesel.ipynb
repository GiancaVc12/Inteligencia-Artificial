```python
import pandas as pd

# Diccionario de meses
meses = {
    'Ene': 1, 'Feb': 2, 'Mar': 3, 'Abr': 4,
    'May': 5, 'Jun': 6, 'Jul': 7, 'Ago': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dic': 12
}

# Leer el archivo
df = pd.read_csv(
    "Mensuales-20260629-111432.csv",
    skiprows=2,
    header=None,
    names=["fecha", "indice_precio_diesel"],
    encoding="cp1252"
)

# Convertir Ene10 -> 2010-01-01
def convertir_fecha(txt):
    mes = meses[txt[:3]]
    anio = 2000 + int(txt[3:])
    return f"{anio}-{mes:02d}-01"

df["fecha"] = df["fecha"].apply(convertir_fecha)

# Redondear a 4 decimales
df["indice_precio_diesel"] = (
    df["indice_precio_diesel"]
    .astype(float)
    .round(4)
)

# Guardar
df.to_csv("indicadores_diesel_limpio.csv", index=False)

print(df.head())
```
